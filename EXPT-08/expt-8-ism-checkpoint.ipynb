```python
import cv2
import string
import os
import matplotlib.pyplot as plt

# Create dictionaries
d = {}
c = {}
for i in range(255):
    d[chr(i)] = i
    c[i] = chr(i)

# Load image
x = cv2.imread(r"C:\Users\admin\Desktop\abc1.jpg")
i, j = x.shape[0], x.shape[1]
print("Image shape:", i, j)

# Encryption
key = input("Enter key to edit (Security Key): ")
text = input("Enter text to hide: ")

kl = 0
tln = len(text)
z = 0
n = 0
m = 0
l = len(text)

for i in range(l):
    x[n, m, z] = d[text[i]] ^ d[key[kl]]
    n = n + 1
    m = m + 1
    m = (m + 1) % 3
    kl = (kl + 1) % len(key)

cv2.imwrite("encrypted_img.jpg", x)

# Show encrypted image inside notebook
plt.imshow(cv2.cvtColor(x, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

print("Data Hiding in Image completed successfully.")

# Decryption
kl = 0
tln = len(text)
z = 0
n = 0
m = 0

ch = int(input("\nEnter 1 to extract data from Image : "))

if ch == 1:
    key1 = input("\n\nRe-enter key to extract text : ")
    decrypt = ""

    if key == key1:
        for i in range(l):
            decrypt += c[x[n, m, z] ^ d[key[kl]]]
            n = n + 1
            m = m + 1
            m = (m + 1) % 3
            kl = (kl + 1) % len(key)

        print("Encrypted text was:", decrypt)
    else:
        print("Key doesn't match.")
else:
    print("Thank you. EXITING.")
```
